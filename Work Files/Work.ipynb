```python
# Calculate descriptive statistics for Overall Engagement and Job Satisfaction
engagement_stats = df['Overall Engagement'].describe()
satisfaction_stats = df['Job Satisfaction'].describe()

# Calculate mode separately, as it's not included in describe()
engagement_mode = df['Overall Engagement'].mode()[0]
satisfaction_mode = df['Job Satisfaction'].mode()[0]

# Display the results
print("Overall Engagement Statistics:")
print(engagement_stats)
print(f"Mode: {engagement_mode}")

print("\nJob Satisfaction Statistics:")
print(satisfaction_stats)
print(f"Mode: {satisfaction_mode}")
```

```python
# Calculate descriptive statistics for Overall Engagement and Job Satisfaction
engagement_stats = df['Overall Engagement Numeric'].describe()
satisfaction_stats = df['Job Satisfaction Numeric'].describe()

# Calculate mode separately
engagement_mode = df['Overall Engagement'].mode()[0]
satisfaction_mode = df['Job Satisfaction'].mode()[0]

# Calculate standard deviation
engagement_std = df['Overall Engagement Numeric'].std()
satisfaction_std = df['Job Satisfaction Numeric'].std()

# Display the results
print("Overall Engagement Statistics:")
print(engagement_stats)
print(f"Mode: {engagement_mode}")
print(f"Standard Deviation: {engagement_std:.2f}")

print("\nJob Satisfaction Statistics:")
print(satisfaction_stats)
print(f"Mode: {satisfaction_mode}")
print(f"Standard Deviation: {satisfaction_std:.2f}")
```
